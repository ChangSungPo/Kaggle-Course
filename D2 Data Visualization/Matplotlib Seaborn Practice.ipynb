## Matplotlib

```python
import pandas as pd
import numpy as np

from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# y = sin(x)
x1 = np.linspace(0, 5 * np.pi, 200)
y1 = np.sin(x1)

#  y = cos(x)
x2 = np.linspace(0, 5 * np.pi, 200)
y2 = np.cos(x2)
```

```python
#figsize = (figure_width, figure_height)
plt.figure(figsize = (10, 6), dpi = 100)

for i in range(4):
    plt.subplot(2, 2, i + 1, title = f'ax{i + 1}')
    plt.plot(x1, y1)

#wspace = distance between figure horizontally
#hspace = distance bwtween figure vertically
plt.subplots_adjust(wspace=0.2, hspace=0.4)
plt.savefig('test.png', facecolor='white')
plt.show()
```

## Seaborn

```python
fig = plt.figure(figsize = (10, 5))

points = np.random.randn(200).cumsum()

ax1 = plt.subplot(1, 2, 1, title = 'bar')
sns.histplot(data = points, ax = ax1)

ax2 = plt.subplot(1, 2, 2, title = 'distribution')
sns.histplot(data = points, kde = True, ax = ax2)
# !!!
# sns.displot(data = points, kde = True, ax = ax2)
# !!!

plt.show()
```
